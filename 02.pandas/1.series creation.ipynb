Creating Series

```python
list = [10,20,30]
array = np.array([10,20,30])
dict = {
    "name":"john",
    "age":18
}
labels = ['a','b','c']
```

```python
pd.Series(list)
```

```python
pd.Series(list,index=labels)
```

array series + custom labels

```python
pd.Series(array)
```

```python
pd.Series(list,index=labels)
```
