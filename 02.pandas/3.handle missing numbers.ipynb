```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

finding missing number

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

removing missing number

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

fill missing numbers

```python
df.fillna(0)
```

```python
val = {"A":1,"B":2,"C":3,"D":4}
df.fillna(val)
```
