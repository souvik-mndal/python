```python
df = pd.read_csv("Countries (1).csv")
df
```

```python
df.fillna(0,inplace=True)
df.isna().sum()
```

```python
#which country has the hightest population
df['country'][df['population'] == df['population'].max()]
```

```python
#capital of the highest population country
df['capital_city'][df['population'] == df['population'].max()]
```

```python
#which country is the least population
df['country'][df['population'] == df['population'].min()]
```

```python
#capital of the least population country
df['capital_city'][df['population'] == df['population'].min()]
```

```python
#top 5 countries with highest democratic score
# df['country'][df['democracy_score'].argsort()][::-1].head()]
df['country'][df['democracy_score'].argsort()][::-1].head()
```
