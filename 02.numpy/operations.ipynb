MATRIX MANIPULATION

```python
a1 = np.array([[1,2],[3,4]])
a2 = np.array([[5,6],[7,8]])
```

```python
a1 * a2
```

```python
a1 @ a2
```

```python
np.dot(a1,a2)
```

```python
a1.T
```

STACKING OF ARRAY

```python
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
```

```python
np.vstack((a1,a2))
```

```python
np.hstack((a1,a2))
```

```python
np.column_stack((a1,a2))
```
